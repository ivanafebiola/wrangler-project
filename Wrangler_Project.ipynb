# Bag 6: Mapping Kategori

## Mapping kategori experience_level

```python
df2['experience_level'].unique()
```

* SE = SENIOR
* MI = MIDDLE
* EN = ENTRY
* EX = EXECUTIVE

```python
# Buat dictionary
dict_exp_level = {
    'SE': 'SENIOR',
    'MI': 'MIDDLE',
    'EN': 'ENTRY',
    'EX': 'EXECUTIVE'
}
```

```python
df2['experience_level'] = (df2['experience_level'].map(dict_exp_level))
```

```python
df2['experience_level'].unique()
```

## Mapping kategori employment_type

```python
df2['employment_type'].unique()
```

```python
dict_emp_type = {
    'FT': 'FULL TIME',
    'CT': 'CONTRACT',
    'FL': 'FREELANCE',
    'PT': 'PART TIME'
}
```

```python
df2['employment_type'] = (df2['employment_type'].map(dict_emp_type))
```

```python
df2['employment_type'].unique()
```

## Mapping kategori remote_ratio

```python
df2['remote_ratio'].unique()
```

```python
dict_remote_ratio = {
    0 : 'ONSITE',
    50 : 'HYBRID',
    100 : 'REMOTE'
}
```

```python
# Membuat kolom baru kategori_remote
df2['remote_category'] = (df['remote_ratio'].map(dict_remote_ratio))
df2
```

# Bag 8: Grouping Data

Pivot table (ga ada jadi kolom)

Berapakah rata-rata gaji berdasarkan experience_level?

```python
df2.groupby('experience_level')['salary'].mean()
```

Berapa gaji maksimum berdasarkan employment_type?

```python
df2.groupby('employment_type')['salary'].max()
```

Contoh groupby gabungan

```python
df2.groupby(['experience_level','employment_type'])[['salary','salary_in_usd']].mean()
```

Berapa maximum salary dan salary_in_usd per company_size dan remote_ratio

```python
df2.groupby(['company_size','remote_ratio'])[['salary','salary_in_usd']].max()
```

```python
# Mau format tabel (tidak di merge)
df2.groupby(['company_size','remote_ratio'])[['salary','salary_in_usd']].max().reset_index()
```

```python
df.groupby('experience_level').agg({
    'salary': 'max',
    'work_year': 'mean'
})
```

Tampilkan data per job_title, tampilkan rata-rata, total, min dari salary_in_usd, dan jumlah (count/size) dari experience_level

```python
df2.groupby('job_title').agg({
    'salary_in_usd': ['mean', 'count', 'min'],
    'experience_level': 'count'
})
```
